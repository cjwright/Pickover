```python
from numpy import random
import math 
import matplotlib.pyplot as plt

scatter = plt.figure()
ax = scatter.add_subplot(projection='3d')

count = 200
arr = random.random(count)

#print(arr)
for i in range(count-2):
    theta=2*math.pi*arr[i]
    phi=math.pi*arr[i+1]
    r=math.pi*arr[i+2]

    x=r*math.sin(theta)*math.cos(phi)
    y=r*math.sin(theta)*math.sin(phi)
    z=r*math.cos(theta)
#    print(theta,phi,r,x,y,z)
    ax.scatter(x, y, z, "o")

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
