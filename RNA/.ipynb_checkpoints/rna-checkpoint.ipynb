```python
import matplotlib.pyplot as plt
from random import randrange

# Declare number of bases 
#
bases = 80 


# Function to initialize sequence with random bases 
# 

def init_sequence(a):
    
    # Initialize sequence to all 0 
    #
    a = [randrange(4) for i in range(bases)]
    
    # Generate random sequence of bases 
    print(a)
    return(a)

# Initialize sequence 
#

sequence=init_sequence(bases)


# Display initial sequence 
#

x=0
    
for num in range(bases):
      if  sequence[num]== 0:
          plt.plot(x, 1, 'r.')
      if  sequence[num]== 1:
          plt.plot(x, 1, 'g.')
      if  sequence[num]== 2:
           plt.plot(x, 1, 'b.')
      if  sequence[num]== 3:
           plt.plot(x, 1, 'y.')
    
      x=x+2
      plt.axis((0, bases*4, 0, 4))
plt.show()

```
